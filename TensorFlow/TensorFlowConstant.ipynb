```
import tensorflow as tf
```

```
con1 = tf.constant(1)
con2 = tf.constant(2.3)
con3 = tf.constant("Saurabh")
con4 = tf.constant(True)
```

```
print(con1, con2 , con3, con4)
```

```
session = tf.Session()
session.run([con1, con2, con3, con4])
```

```
add = tf.constant(20) + tf.constant(30)
mul = tf.constant(20) * tf.constant(30)
session.run([add, mul])
```

```
Addition = tf.constant([1,2,3,4]) + tf.constant([4,3,2,1])
Product = tf.constant([1,2,3,4]) * tf.constant([4,3,2,1])

session.run([Addition, Product])
```
