```
import tensorflow as tf
```

```
a = tf.placeholder(tf.int32)
b = a * 2
```

```
session = tf.Session()
session.run(b, feed_dict={a: 4})
```

```
session.run(b, feed_dict={a: [1,2,3,4]})
```

```
str_name = tf.placeholder(tf.string)
my_name = "My Name is " + str_name
```

```
session.run(my_name, feed_dict={str_name: ["saurabh", "Sharma"]})
```
